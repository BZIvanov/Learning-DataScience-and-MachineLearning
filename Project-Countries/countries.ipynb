### Fix missing data

Let's see for which columns how much data we are missing

```python
df.isnull().sum()
```

Let's see which are the countries for which we are missing Agriculture data.

```python
df[df['Agriculture'].isnull()]['Country']
```

Because most of the countries which are missing Agriculture are small islands we will just use zero as value for them.

```python
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

For Climate we will use the mean value for the region of the country.

```python
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

For Literacy we will also use the mean value for the Region.

```python
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

Let's check again what is left to be fixed.

```python
df.isnull().sum()
```

There are just a few N/A records so for simplicity we will just drop them.

```python
df = df.dropna()
```

### Data feature preparation

The Country column is still a unique identifier string, so it won't be useful for clustering, since it is unique for each point so let's remove it. Let's also get dummy values for the string value columns.

```python
X = df.drop("Country", axis=1)
X = pd.get_dummies(X)
X.head()
```

### Scaling

Let's scale the data, because the values ranges for the different columns are very different.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
scaled_X
```

### Creating and Fitting Kmeans Model

```python
from sklearn.cluster import KMeans
```

```python
# ignore Windows specific warning for environment variable
import warnings
warnings.filterwarnings('ignore')

sum_squared_distances = []

for k in range(2, 30):
    
    model = KMeans(n_clusters=k, n_init=10)
    model.fit(scaled_X)
    sum_squared_distances.append(model.inertia_)
```

```python
plt.plot(range(2, 30), sum_squared_distances, 'o--')
plt.xlabel("K Value")
plt.ylabel("Sum of Squared Distances")
```

Let's choose K=4

```python
pd.Series(sum_squared_distances).diff().plot(kind='bar')
```

### Model Interpretation

After we selected 4 to be our K value, let's use it below

```python
model = KMeans(n_clusters=4, n_init=10)
model.fit(scaled_X)
```

```python
model.labels_
```

```python
X['K=4 Clusters'] = model.labels_
```

```python
X.corr()['K=4 Clusters'].sort_values()
```

```python
X.corr()['K=4 Clusters'].sort_values().plot(kind='bar')
```
